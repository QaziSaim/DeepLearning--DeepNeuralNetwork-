```python
import re
def clean_text(text):
    text = re.sub(r'[^A-Za-z0-9\s]', '', text)       # remove special characters
    text = re.sub(r'\s+', ' ', text).strip()         # remove newlines/tabs/extra spaces
    return text
```

```python
df['text'] = df['text'].apply(clean_text)
```

```python
df['text'] = df['text'].apply(lambda n:n.lower())
```
