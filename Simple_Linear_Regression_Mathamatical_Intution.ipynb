```python
X = [1, 2, 3, 4, 5]
Y = [2, 4, 5, 4, 5]
```

```python
print(X,Y)
```

```python
def mean(values):
  return sum(values)/len(values)
```

```python
def linear_regression(X, Y):
  n = len(X)
  sum_x = sum(X)
  sum_y = sum(Y)
  sum_xy = sum(x* y for x,y in zip(X,Y))
  sum_x2 = sum(x**2 for x in X)
  m = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x ** 2)
  c = (sum_y - m * sum_x) / n
  return m,c
```

```python
def predict(x,m,c):
  return m * x + c
```

```python
m, c = linear_regression(X, Y)
```

```python
x_new = 6
y_pred = [predict(x, m, c) for x in X]
print(x_new,y_pred)
```
